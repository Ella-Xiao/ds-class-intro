```python
# 引入pandas,打开文件
import pandas as pd
pd.set_option('display.max_columns',100)

college = pd.read_csv('data/college.csv')
```

```python
college.head()
```

# Missing values

```python
# variable(columns) missing rate --- notna()
college['satvrmid'].notna().mean()
```

```python
# 看几行几列---
college.shape
```

```python
# drop missing rate后剩下几行几列 & 实际计算non-NaN value ---- dropna()
college.dropna().shape
```

```python
# df.dropna(subset=['colunmns name 1, 2,3'])
# Define in which columns to look for missing values.
college.dropna(subset=['satvrmid']).shape
```

```python
# axis : {0 or 'index', 1 or 'columns'}, default 0
# Determine if rows or columns which contain missing values are removed.
college.dropna(axis=1)
```

```python
# df.fillna --填充空值，backfill: use next valid observation to fill gap.
# 另外一种method, ffill: propagate last valid observation forward to next valid backfill
college.fillna(method='backfill').head()
```

```python
# 计算某个columns的median()
stavrmid_med=college['satvrmid'].median()
```

```python
# 将median的值补充到空值上，根据正态分布的规律： fillna(value)
#values specifying which value to use for each index (for a Series) or column (for a DataFrame)
college['satvrmid'].fillna(value=stavrmid_med)
```

# Arithmetic and Comparison Operators

```python
# method跟operator 在一些范围内是一样的
ugds=college['ugds'] # 将series赋值
ugds_operator=ugds+100
ugds_method=ugds.add(100)
ugds_operator.equals(ugds_method)
```

# Operators and methods may perform differently on your data

```python
# df.set_index 将columns设置为index
college_idx=college.set_index('instnm')
college_race=college_idx.loc[:,'ugds_white':'ugds_unkn'] # 【：】表示默认index, selected columns
# college_race属于DataFrane
college_race.head()
```

```python
type(college_race)
```

```python
# 计算 Data Frame * Series 
# series的index必须和Frame的row index对应（也就是必须为行而不能为列），否则计算会出现奇怪的结果
```

# Use the axis parameter of the mul method

```python
#计算ugds---- series
ugds = college_idx['ugds']
ugds.head()
```

```python
type(ugds)
```

```python
type(college_race)
```

```python
#instead of aligning the columns of a DataFrame to the index of a Series, 
# we can align the index of a DataFrame to the index of a Series
#难题
df_correct=college_race.mul(ugds,axis='index').round(0)
df_correct.head()
```
