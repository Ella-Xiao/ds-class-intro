```python
df=pd.read_csv('data/sample_data.csv')
name=df['name'].copy() # 建立一个copy件，在此文件做出修改
name.loc[0]='Emma'
name #得到了copy修改后的数据，而且没有出现warning signals.
```

# setting with copy warning

```python
# filer /condition indexing --- 直接可以筛选条件
df[df['age']<18]
```

```python
# 后面加上一个新的series，对该series 进行assignment
df[df['age']<18]['score']=0
```

```python
df  # 发现对value assignment 没有产生任何变化  --- new DF is never assigned to a varible
```

```python
# correct way  ?? 感觉比较抽象--- 利用df.loc对series 进行赋值
# df.loc : Set value for all items matching the list of labels 
# Set value for rows matching callable condition

df.loc[df['age']<18,'score']=0
df
```

```python
df_new=df[['state','count']]  # 变为 DataFrame
df_new
```

```python
# 在原有df的基础上进行修改, 又会出现 warning
df_new['count']=0
```

```python
df_new
```

```python
df 
```

```python
# 出现warning --- 使用copy()
```

```python
df_new=df[['state','count']].copy()
df_new['count']=0  # 对某一个series进行assignment
df_new # 加了copy(), 不会出现warning
```
